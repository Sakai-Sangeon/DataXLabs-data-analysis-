```python
import pandas as pd

# 파일이 Colab의 /content 폴더에 올라감
file_path = '/content/Superstore(Task 5).csv'

df = pd.read_csv(file_path, encoding='cp1252')
print("✅ 파일 불러오기 성공")
print("데이터 크기:", df.shape)
print("컬럼 이름:", list(df.columns)[:5])
```

```python
df.head()
```

```python
# 데이터 정보 확인
df.info()

# 수치형 데이터의 기본 통계 요약
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Numeric columns 확인
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Numeric columns:", num_cols)

# 히스토그램 (Histogram)
for col in num_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
print("Categorical columns:", cat_cols)

for col in ['Category', 'Sub-Category', 'Region', 'Segment', 'Ship Mode']:
    if col in df.columns:
        plt.figure(figsize=(6,4))
        sns.countplot(y=col, data=df, order=df[col].value_counts().index, palette='Set2')
        plt.title(f"Count of {col}")
        plt.show()
```

### Observation (Univariate Analysis)
- Sales and Profit are right-skewed.
- Quantity distribution is nearly uniform.
- Most customers are in the Consumer segment.

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency, pearsonr

sns.set(style="whitegrid")
%matplotlib inline

# CSV 파일 읽기 (인코딩 문제 방지)
df = pd.read_csv('Superstore(Task 5).csv', encoding='latin1')

# 데이터 확인
df.head()
df.info()
```

```python
# 변수 이름 예시: Sales, Profit
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs Profit')
plt.show()

# 상관계수 계산
corr, _ = pearsonr(df['Sales'], df['Profit'])
print(f"Correlation between Sales and Profit: {corr:.2f}")
```

```python
# 숫자형 컬럼만 선택
numeric_df = df.select_dtypes(include=np.number)

# 상관계수 계산
corr_matrix = numeric_df.corr()

# 히트맵 시각화
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap (Numeric Columns Only)')
plt.show()

```

```python
cross_tab = pd.crosstab(df['Category'], df['Region'])
print(cross_tab)

sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Category vs Region')
plt.show()
```

```python
chi2, p, dof, expected = chi2_contingency(cross_tab)
print(f"Chi-square test statistic: {chi2:.2f}, p-value: {p:.4f}")
```
